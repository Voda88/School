```python
temperatures_for_debye=np.linspace(298.15,2000,1000)
dimensionless_temperatures=temperatures_for_debye/743

debye_integral=lambda x: (np.exp(x)*x**4)/((np.exp(x)-1)**2)
heat_capacity_integrals=np.zeros(len(dimensionless_temperatures))

for i,item in enumerate(dimensionless_temperatures):
    heat_capacity_integrals[i]=sp.integrate.quad(debye_integral,0,1/item)[0]

heat_capacities_per_mole=np.zeros(len(heat_capacity_integrals))
                                  
for i in range(0,len(heat_capacities_per_mole)):
    heat_capacities_per_mole[i]=9*scipy.constants.k*scipy.constants.N_A*(dimensionless_temperatures[i])**3*heat_capacity_integrals[i]

def Calculate_cv(T):
    TD=T/743
    a=sp.integrate.quad(debye_integral,0,1/TD)[0]
    return 9*scipy.constants.k*scipy.constants.N_A*TD**3*a/0.0403

    
```
