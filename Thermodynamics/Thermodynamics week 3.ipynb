### Problem 1.

For ideal gas undertaking the following processes (Carnot cycle), show 

$$ \eta = \frac{Q_2+Q_1}{Q_2} = \frac{T_2-T_1}{T_2}$$

### Solution:

The carnot cycle requires the heat engine to return to its original state after a full cycle. This means that the total change of internal energy of the system $(\Delta U)$ after a complete cycle is equal to 0. In other terms, the sum of heat received $ (Q_2) $, heat discharged $(Q_1)$ and work done by the system $(W)$ is equal to zero:

$$ \Delta U = 0 = Q_2+Q_1+W $$

Rearranging for work:

$$-W = Q_2 + Q_1$$

which means that the work done by the system is equal in magnitude to the sum of the heat received and the heat discharged (heat received is positive while heat discharged is negative).

For a heat engine, efficiency is given by the ratio of work done by the engine $(W) $ and the heat received by the engine $(Q_2)$ i.e. how large a fraction of the heat received can be used for work. A negative sign is added to make the efficiency non-negative (since work done *by* the system is negative):

$$ \eta = -\frac{W}{Q_2} $$

Using the definition for work above:

$$ \eta = \frac{Q_2+Q_1}{Q_2} $$

The Carnot cycle includes 4 stages: isothermal expansion, adiabatic expansion, isothermal compression and adiabatic compression. The system receives all its heat during isothermal expansion and discharges all of the heat it was unable to use for work during isothermal compression.


Therefore, for an ideal gas, $ Q_2 $ is given by:
$$ RT_2ln\frac{V_2}{V_1} $$

and $ Q_1 $ is given by:

$$ RT_1ln\frac{V_4}{V_3} $$

For the carnot cycle, the ratio of the volumes before and after isothermal expansion/compression must remain constant:

$$ \frac{V_2}{V_1} = \frac{V_3}{V_4} $$

It follows that

$$ Q_1+Q_2 = RT_2ln\frac{V_2}{V_1} + RT_1ln\frac{V_4}{V_3} = RT_2ln\frac{V_2}{V_1} - RT_1ln\frac{V_2}{V_1} = R(T_2 - T_1)ln\frac{V_2}{V_1} $$

The expression for efficiency can be expressed as:

$$ \eta = \frac{Q_2+Q_1}{Q_2}= \frac{R(T_2 - T_1)ln\frac{V_2}{V_1}}{RT_2ln\frac{V_2}{V_1}} = \frac{T_2-T_1}{T_2}$$

$$ P_3=P_2 \frac{T_1}{T_2}^{7/2} $$

Calculating:

$$ P_3=3 \frac{315}{700}^{7/2} = 0.061 $$

Using:

$$ P_4 = P_3\frac{P_1}{P_2}   $$

rearranging:

$$ P_4 = P_3\frac{P_1}{P_2} =0.061 \cdot \frac{3}{10} = 0.203 $$
